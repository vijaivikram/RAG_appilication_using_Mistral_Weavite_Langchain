```python
from google.colab import userdata
WEAVIATE_CLUSTER = userdata.get('CLUSTER_URL')
WEAVIATE_API_KEY = userdata.get('CLUSTER_API')
```

```python
from langchain.vectorstores import Weaviate
import weaviate

client = weaviate.Client(url=WEAVIATE_CLUSTER, auth_client_secret=weaviate.AuthApiKey(WEAVIATE_API_KEY))
```

```python
# fixing unicode error in google colab
import locale
locale.getpreferredencoding = lambda: "UTF-8"
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
embedding_model_name = 'sentence-transformers/all-mpnet-base-v2'
model_kwargs={'device':'cuda'}
embeddings = HuggingFaceEmbeddings(model_name=embedding_model_name,model_kwargs=model_kwargs)
```

```python
from langchain.document_loaders import PyPDFLoader
loader = PyPDFLoader('/content/rag.pdf')
pages = loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
docs = text_splitter.split_documents(pages)
```

```python
vector_db = Weaviate.from_documents(docs,embeddings, client=client, by_text=False)
```

```python
from langchain.prompts import ChatPromptTemplate

template = '''You are the most intelligent and helpful assistant for question-answering tasks. Use the following pieces of retrieved context to answer the question.
If you don't know the answer, say you don't know. Try to keep the answer concise.
Context : {context}
Answer : '''

prompt = ChatPromptTemplate.from_template(template)
```

```python
from langchain import HuggingFaceHub

from google.colab import userdata
huggingface_api_key = userdata.get('HUGGINGFACE_API')
```

```python
model = HuggingFaceHub(huggingfacehub_api_token=huggingface_api_key,repo_id='mistralai/Mistral-7B-Instruct-v0.2',
                       model_kwargs={'temperature':1,'max_length':2000})
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser

output_parser = StrOutputParser()
retriever = vector_db.as_retriever()
```

```python
rag_chain = ({'context':retriever, 'question':RunnablePassthrough()}
             | prompt | model | output_parser)
```

```python
print(rag_chain.invoke('Explain rag system'))
```
